```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps


from Config import gkey
```

```python
Coronavirus_data = pd.read_csv("us_healthcare_capacity-county-CovidCareMap.csv")
Coronavirus_data.head()   
```

```python
cleaned_data = Coronavirus_data.drop(['Population', 'Population (20+)','Staffed ICU Beds [Per 1000 People]', 'Licensed All Beds [Per 1000 People]','Licensed All Beds [Per 1000 Adults (20+)]','Licensed All Beds [Per 1000 Elderly (65+)]','ICU Bed Source', 'ICU Bed Source Last Updated'], axis=1)
cleaned_data
```

```python
extra_clean_data = cleaned_data.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
extra_clean_data
```

```python
us_coord = pd.read_csv("coordinates.csv")
us_coord.head()
```

```python
merged_data = pd.merge(extra_clean_data, us_coord, how='outer', on=['State', 'State'])
merged_data
```

```python
# number of countys
Counties = extra_clean_data['fips_code'].count()
Counties
```

```python
#hospital beds per county in each state
hospital_beds = extra_clean_data.groupby(['State']).sum()['Staffed All Beds']
hospital_beds
```

```python
#occupancy rate in hospitals per county in each state
occupancy_rate = extra_clean_data.groupby(['State']).agg({'All Bed Occupancy Rate' : 'sum'})
occupancy_rate

# set "State" as an index column for us_coord
us_coord = us_coord.set_index("State")

# merge both dfs
state_occupancy_coord_merged_df = occupancy_rate.merge(us_coord, left_index=True, right_index=True)
state_occupancy_coord_merged_df.head()
```

```python
#Configure gmaps.
gmaps.configure(api_key = gkey)
#Use the Lat and Lng as locations and State as the weight.
locations = state_occupancy_coord_merged_df[["lat", "lng"]]
#Add Heatmap layer to map.
State = state_occupancy_coord_merged_df["All Bed Occupancy Rate"]
```

```python
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

params = {"type" : "lodging",
          "radius" : 50,
          "key" : gkey}
```

```python
info_box_template = """
<dl>
<dt>State</dt><dd>{state}</dd>
<dt>Occupancy</dt><dd>{occupancy}</dd>
</dl>
"""

state_info = []
for state, occupancy in state_occupancy_coord_merged_df.iterrows():
    state_info.append(info_box_template.format(state=state, occupancy=round(occupancy["All Bed Occupancy Rate"], 2)))

# Display the map
# map for locations and Occupancy Rate
fig = gmaps.figure(center=(43.075970, -107.290283), zoom_level=4)

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations,
                                 weights = State,
                                 dissipating=False,
                                 max_intensity=np.max(State),
                                 point_radius=4
                                )

markers = gmaps.marker_layer(locations=state_occupancy_coord_merged_df[['lat', 'lng']],
                             info_box_content=state_info)



# Add layer
fig.add_layer(heat_layer)
fig.add_layer(markers)

# Display figure
fig
#plt.savefig("All Beds Occupancy Rates.png")
```

```python
#icu occupancy rate in hospitals per county in each state
icu_occupancy_rate = extra_clean_data.groupby(['State']).agg({'ICU Bed Occupancy Rate' : 'sum'})
icu_occupancy_rate

# set "State" as an index column for us_coord
#us_coord = us_coord.set_index("State")

# merge both dfs
icu_state_occupancy_coord_merged_df = icu_occupancy_rate.merge(us_coord, left_index=True, right_index=True)
icu_state_occupancy_coord_merged_df.head()
```

```python
#Configure gmaps.
gmaps.configure(api_key = gkey)
#Use the Lat and Lng as locations and State as the weight.
locations = icu_state_occupancy_coord_merged_df[["lat", "lng"]]
#Add Heatmap layer to map.
State = icu_state_occupancy_coord_merged_df["ICU Bed Occupancy Rate"]
```

```python
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

params = {"type" : "lodging",
          "radius" : 50,
          "key" : gkey}
```

```python
info_box_template = """
<dl>
<dt>State</dt><dd>{state}</dd>
<dt>ICUOccupancy</dt><dd>{ICUoccupancy}</dd>
</dl>
"""

state_info = []
for state, ICUoccupancy in icu_state_occupancy_coord_merged_df.iterrows():
    state_info.append(info_box_template.format(state=state, ICUoccupancy=round(ICUoccupancy["ICU Bed Occupancy Rate"], 2)))

# Display the map
# map for locations and ICU Occupancy Rate
fig = gmaps.figure(center=(34.799999,-92.199997), zoom_level=4)

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations,
                                 weights = State,
                                 dissipating=False,
                                 max_intensity=np.max(State),
                                 point_radius=4
                                )

markers = gmaps.marker_layer(locations=icu_state_occupancy_coord_merged_df[['lat', 'lng']],
                             info_box_content=state_info)



# Add layer
fig.add_layer(heat_layer)
fig.add_layer(markers)

# Display figure
fig
```
